## PARTE 3 ‚Äì Encapsulamento e Abstra√ß√£o

**Tarefas:**
- Tornar pelo menos 1 atributo da classe principal privado (`__atributo`).
- Criar m√©todos `get` e `set` para o atributo privado.
- Implementar um m√©todo que contenha l√≥gica interna (ex: c√°lculo, verifica√ß√£o).

**C√≥digo:**

```python
from datetime import datetime

class Viagem_Danilo:
    def __init__(self, destino, data, status="Planejada"):
        self.destino = destino
        self.data = data
        self.__status = status  # Atributo privado

    def get_status(self):
        return self.__status

    def set_status(self, novo_status):
        if novo_status in ["Planejada", "Confirmada", "Cancelada"]:
            self.__status = novo_status
        else:
            print("Status inv√°lido!")

    # Novo m√©todo com l√≥gica interna
    def dias_para_viagem(self):
        data_viagem = datetime.strptime(self.data, "%d/%m/%Y")
        hoje = datetime.today()
        diferenca = (data_viagem - hoje).days
        return diferenca

    def __str__(self):
        return f"Destino: {self.destino}\nData: {self.data}\nStatus: {self.__status}"


# Testando
viagem1 = Viagem_Danilo("Rio de Janeiro", "10/07/2025")
print(viagem1)

print("\nStatus atual:", viagem1.get_status())

viagem1.set_status("Confirmada")
print("Novo status:", viagem1.get_status())

dias = viagem1.dias_para_viagem()
if dias > 0:
    print(f"Faltam {dias} dias para a viagem.")
elif dias == 0:
    print("A viagem √© hoje!")
```

```python
# Importa a biblioteca mysql.connector para conectar ao banco de dados MySQL
import mysql.connector

# Importa a fun√ß√£o datetime para manipular datas
from datetime import datetime


# === CLASSE PRINCIPAL (POO) ===
class Viagem_Danilo:
    def __init__(self, destino, data, status="Planejada"):
        self.destino = destino         # Atributo p√∫blico
        self.data = data               # Atributo p√∫blico
        self.__status = status         # Atributo privado (encapsulamento)

    # Getter para acessar o status (encapsulamento)
    def get_status(self):
        return self.__status

    # Setter para alterar o status com verifica√ß√£o
    def set_status(self, novo_status):
        if novo_status in ["Planejada", "Confirmada", "Cancelada"]:
            self.__status = novo_status

    # Calcula quantos dias faltam at√© a viagem (l√≥gica interna)
    def dias_para_viagem(self):
        data_viagem = datetime.strptime(self.data, "%d/%m/%Y")  # Converte string em data
        hoje = datetime.today()                                 # Pega a data atual
        return (data_viagem - hoje).days                        # Retorna a diferen√ßa em dias

    # M√©todo especial que define como o objeto ser√° impresso (polimorfismo)
    def __str__(self):
        dias = self.dias_para_viagem()
        return f"Destino: {self.destino}\nData: {self.data}\nStatus: {self.__status}\nDias restantes: {dias}"


# === SUBCLASSE COM HERAN√áA E POLIMORFISMO ===
class ViagemInternacional_Danilo(Viagem_Danilo):
    def __init__(self, destino, data, passaporte, status="Planejada"):
        super().__init__(destino, data, status)   # Chama o construtor da classe m√£e
        self.passaporte = passaporte              # Novo atributo exclusivo

    # M√©todo exclusivo da subclasse para verificar documenta√ß√£o
    def verificar_documentos_Danilo(self):
        if self.passaporte:
            return "Documentos OK para viagem internacional."
        else:
            return "Passaporte ausente! Regularize antes da viagem."

    # Sobrescreve __str__ da classe m√£e (polimorfismo)
    def __str__(self):
        base_info = super().__str__()  # Usa o m√©todo da classe m√£e
        return f"{base_info}\nPassaporte: {'Sim' if self.passaporte else 'N√£o'}"


# === FUN√á√ïES PARA BANCO DE DADOS ===

# Cria o banco de dados caso ele ainda n√£o exista
def criar_banco_se_nao_existir():
    conexao = mysql.connector.connect(
        host="localhost",
        user="root",     # usu√°rio
        password=""      # senha
    )
    cursor = conexao.cursor()
    cursor.execute("CREATE DATABASE IF NOT EXISTS viagens_db")
    conexao.close()

# Faz conex√£o com o banco de dados
def conectar():
    return mysql.connector.connect(
        host="localhost",
        user="root",
        password="",
        database="viagens_db"
    )

# Cria a tabela de viagens se ainda n√£o existir
def criar_tabela():
    conn = conectar()
    cursor = conn.cursor()
    cursor.execute("""
        CREATE TABLE IF NOT EXISTS viagens (
            id INT AUTO_INCREMENT PRIMARY KEY,
            destino VARCHAR(100),
            data VARCHAR(20),
            status VARCHAR(20)
        )
    """)
    conn.commit()
    conn.close()

# Insere uma nova viagem no banco de dados
def inserir_viagem(viagem):
    conn = conectar()
    cursor = conn.cursor()
    sql = "INSERT INTO viagens (destino, data, status) VALUES (%s, %s, %s)"
    val = (viagem.destino, viagem.data, viagem.get_status())  # Usa o getter
    cursor.execute(sql, val)
    conn.commit()
    conn.close()

# Lista todas as viagens cadastradas no banco
def listar_viagens():
    conn = conectar()
    cursor = conn.cursor()
    cursor.execute("SELECT * FROM viagens")
    resultados = cursor.fetchall()
    conn.close()

    # Exibe as viagens
    if not resultados:
        print("\nüì≠ Nenhuma viagem cadastrada.")
    else:
        print("\nüìã Viagens cadastradas:")
        for v in resultados:
            print(f"ID: {v[0]} | Destino: {v[1]} | Data: {v[2]} | Status: {v[3]}")

# Exclui uma viagem com base no ID fornecido
def excluir_viagem_por_id(id_viagem):
    conn = conectar()
    cursor = conn.cursor()
    cursor.execute("DELETE FROM viagens WHERE id = %s", (id_viagem,))
    conn.commit()
    conn.close()
    print(f"üóëÔ∏è Viagem com ID {id_viagem} foi exclu√≠da.")


# === EXECU√á√ÉO PRINCIPAL COM MENU ===

# Garante que o banco e a tabela existem antes de usar
criar_banco_se_nao_existir()
criar_tabela()

# Loop principal do menu
while True:
    print("\n=== MENU ===")
    print("1 - Listar viagens")
    print("2 - Excluir viagem por ID")
    print("3 - Cadastrar nova viagem")
    print("4 - Sair")
    opcao = input("Escolha uma op√ß√£o: ").strip()

    # Op√ß√£o: Listar viagens
    if opcao == "1":
        listar_viagens()

    # Op√ß√£o: Excluir por ID
    elif opcao == "2":
        listar_viagens()
        try:
            id_excluir = int(input("Digite o ID da viagem a excluir: "))
            excluir_viagem_por_id(id_excluir)
        except ValueError:
            print("‚ùå ID inv√°lido.")

    # Op√ß√£o: Cadastrar nova viagem
    elif opcao == "3":
        tipo = input("Cadastrar viagem [N]acional ou [I]nternacional? ").strip().upper()
        destino = input("Destino da viagem: ")
        data = input("Data da viagem (DD/MM/AAAA): ")

        # Verifica o tipo e cria o objeto correspondente
        if tipo == "I":
            passaporte_ok = input("Possui passaporte v√°lido? [S/N]: ").strip().upper() == "S"
            viagem = ViagemInternacional_Danilo(destino, data, passaporte_ok)
        else:
            viagem = Viagem_Danilo(destino, data)

        # Mostra dados e salva no banco
        print("\n‚úÖ Viagem cadastrada com sucesso:")
        print(viagem)
        inserir_viagem(viagem)

    # Op√ß√£o: Sair do programa
    elif opcao == "4":
        print("Saindo do sistema. At√© logo!")
        break

    # Op√ß√£o inv√°lida
    else:
        print("‚ùå Op√ß√£o inv√°lida. Tente novamente.")
```
